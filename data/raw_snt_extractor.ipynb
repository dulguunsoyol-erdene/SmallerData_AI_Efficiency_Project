```python
import nltk
import csv
import re

nltk.download('punkt')

file_path = "source/articles.txt"
length_threshold = 6
tolerance = 2

suffix_to_label = {
    "сан": "PAST", "сэн": "PAST", "сон": "PAST", "сөн": "PAST",
    "даг": "HABITUAL", "дэг": "HABITUAL", "дог": "HABITUAL", "дөг": "HABITUAL",
    "на": "FUTURE", "нэ": "FUTURE", "но": "FUTURE", "нө": "FUTURE",
}

def label_sentence(sentence):
    words = sentence.split()
    labels = []
    for i, word in enumerate(words):
        clean = word.rstrip('.,!?;:').lower()
        if i == len(words) - 1:
            matched = False
            for suffix, label in suffix_to_label.items():
                if clean.endswith(suffix):
                    labels.append(label)
                    matched = True
                    break
            if not matched:
                labels.append("OTHER")
        else:
            labels.append("OTHER")
    return " ".join(labels)

with open(file_path, "r", encoding="utf-8") as f:
    text = f.read()

sentences = nltk.sent_tokenize(text)
matches = []

for i, sentence in enumerate(sentences, 1):
    words = nltk.word_tokenize(sentence)
    length = len(words)

    if abs(length - length_threshold) <= tolerance:
        label_str = label_sentence(sentence)
        matches.append([sentence, label_str])

with open("filtered_sentences.csv", "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["sentence", "label"])
    writer.writerows(matches)

print(f"Done. {len(matches)} sentences written to 'filtered_sentences.csv'.")
```
